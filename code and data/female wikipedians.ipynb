```python
female_wikipedians=[]
from requests import get
from bs4 import BeautifulSoup
import re
```

```python

url='https://en.wikipedia.org/w/index.php?title=Category:Female_Wikipedians'
headers = {
    'From': '<jf3105@columbia.edu>'
}

response = get(url, headers=headers)
soup = BeautifulSoup(response.text, 'html.parser')
for each in soup.find_all('a', {'title':re.compile('User')}):
            female_wikipedians.append(each.text) 
female_wikipedians
```

```python
female_wikipedians=[]

html = driver.page_source 
soup = BeautifulSoup(html, 'html.parser')
for each in soup.find_all('a', {'title':re.compile('User')}):
    female_wikipedians.append(each.text) 
            
while True: 
    button = driver.find_element_by_link_text('next page')
    try: 
        button.click() 
        html = driver.page_source 
        soup = BeautifulSoup(html, 'html.parser')
        for each in soup.find_all('a', {'title':re.compile('User')}):
            female_wikipedians.append(each.text) 
            
    except: 
        break
        
female_wikipedians
```

```python
from csv import writer

csv_file = open('female wikipedians.csv', 'w')
csv_writer = writer(csv_file)
csv_writer.writerow(['user','edits'])

for each in female_wikipedians:
    if 'Template' not in each \
and 'Userboxes:' not in each and 'User:Technical' not in each \
and 'Wikipedia:Userboxes' not in each and 'UBX' not in each:
        each=re.sub('User:|User talk:|Usertalk:|Userboxes:|/.*','',each)
        if each not in sample_list:
            csv_writer.writerow([each])
```
